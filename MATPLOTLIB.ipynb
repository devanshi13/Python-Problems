```python
import matplotlib.pyplot as plt
```

```python
#line plot
x = [1,2,3,4,5,6]
y = [4,6,2,4,3,2]
plt.plot(x,y,'blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('plot between x and y')
plt.show()
```

```python
#scatter plot
x1 = [1,2,3,4]
y1 = [1,8,3,2]
plt.scatter(x1,y1,color="red")
plt.xlabel('x')
plt.ylabel('y')
plt.title('plot between x and y')
plt.show()
```

```python
#using legend
x = [1,2,3,4,5,6]
y = [4,6,2,4,3,2]
x1 = [1,2,3,4]
y1 = [1,8,3,2]
plt.plot(x,y,label="plot1")
plt.plot(x1,y1,label="plot2")
plt.xlabel('x')
plt.ylabel('y')
plt.title('plot between x and y')
plt.legend()
plt.show()
```

```python
#bar graph with legend
x = [1,2,3,4]
y = [4,6,2,4]
x1 = [1,2,3,4]
y1 = [1,8,3,2]
plt.bar(x,y,label="plot1")
plt.bar(x1,y1,label="plot2")
plt.xlabel('x')
plt.ylabel('y')
plt.title('plot between x and y')
plt.legend()
plt.show()
```

```python
x=[1,2,3,4]
labels=['a','b','c','d']
plt.pie(x,labels=labels)
plt.legend()
plt.show()
```

```python
#multiplots on same canvas
plt.subplot(1,2,1)
x = [1,2,3,4,5,6]
y = [4,6,2,4,3,2]
plt.plot(x,y,'blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('plot between x and y')


plt.subplot(1,2,2)
x1 = [1,2,3,4]
y = [10,9,8,7]
plt.plot(x1,y1,'red')
plt.xlabel('x1')
plt.ylabel('y1')
plt.title('plot between x1 and y1')

plt.legend()
plt.show()
```
